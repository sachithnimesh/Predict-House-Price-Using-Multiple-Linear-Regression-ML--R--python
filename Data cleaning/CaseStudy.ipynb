Mutate the variables. Discretize data in variables related to vaccine coverage

```python
df.info()
```

```python
def g(x):
  if x < 90:
    return '<90 covered'
  else:
    return '>=90 covered'

df['Polio'] = df['Polio'].apply(g)
df['Diphtheria ']=df['Diphtheria '].apply(g)
```

```python
df['Hepatitis B']=df['Hepatitis B'].apply(g)
```
